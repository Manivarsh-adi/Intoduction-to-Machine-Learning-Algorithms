# Considerations
#**Let us take a data which follows a Gaussian Distribution**

```python
import numpy as np
import pandas as pd
from sklearn import datasets
iris=datasets.load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['target']=iris.target
df.head()
```

```python
def PDF_data(x):
    mean_std=((x-x.mean())**2)/(2*(x.std()*x.std()))
    Exponential_Value=np.exp(-mean_std)
    first=(1/(np.sqrt(2)*(22/7)*x.std()))
    pdf=first*Exponential_Value
    return pdf
#data=df[df['target']==0]['sepal length (cm)'].values
```

```python
import matplotlib.pyplot as plt
for i in df.columns[:-1]:
    data=df[i].values
    pdf=PDF_data(data)
    plt.plot(data,pdf)
    plt.xlabel('data')
    plt.ylabel('PDF')
    plt.title(i)
    plt.show()
```

# Q-Q plot

```python
import pylab
from scipy import stats
```

```python
for i in df.columns[:-1]:
    stats.probplot(df[i].values,dist='norm',plot=pylab)
    pylab.show()
```

# Conclusion
#**from Q-Q plot sepal length and sepal width follows Gaussian Distribution (Not Exactly)**
#**But i can assume that sepal length and sepal width follows Gaussian Distribution**
#**from first two plots of probplot sepal length and sepal width tends to follow Guassian Distribution**
#**If i assume or consider Sepal Length and Sepal Width follow Gaussian Distribution then i can depect the information from 68-95-99.7 rule**

# Observations of Sepal Length Using 68-95-99.7 rule


```python
mean_sepal_length=df['sepal length (cm)'].mean()
std_sepal_length=df['sepal length (cm)'].std()

# 68% of values of sepal length  lies in [mean-1*(std), mean+1*(std)]

_68_percent_values=[mean_sepal_length-1*std_sepal_length,mean_sepal_length+1*std_sepal_length]
print('68% of values of sepal length lies in Intervel of',_68_percent_values)

# 95% of values sepal length lies in [mean-2*(std),mean+2*(std)]

_95_percent_values=[mean_sepal_length-2*std_sepal_length,mean_sepal_length+2*std_sepal_length]
print('95% of values of sepal length lies in Intervel of',_95_percent_values)

#99.7 of values sepal length lies in [mean-3*(std),mean+2*(std)]

_99_percent_values=[mean_sepal_length-3*std_sepal_length,mean_sepal_length+3*std_sepal_length]
print('99% of values of sepal length lies in Intervel of',_99_percent_values)
```

# Observation of Sepal width Using 68-95-99.7 rule

```python
mean_sepal_width=df['sepal width (cm)'].mean()
std_sepal_width=df['sepal width (cm)'].std()

#68% of the values lie in [mean-1*std,mean+1*std]

_68_percent_values=[mean_sepal_width-1*std_sepal_width,mean_sepal_width+1*std_sepal_width]
print('68% of sepal width values lie in Interval of',_68_percent_values)

#95% of the values lie in[mean-2*std,mean+2*std]

_95_percent_values=[mean_sepal_width-2*std_sepal_width,mean_sepal_width+2*std_sepal_width]
print('95% of sepal width values lie in Intervel of',_95_percent_values)

#99.7 percent values lie in [mean-3*std,mean+3*std]

_99_percent_values=[mean_sepal_width+3*std_sepal_width,mean_sepal_width+3*std_sepal_width]
print('99% of sepal width values lie in Intervel of',_99_percent_values)
```
