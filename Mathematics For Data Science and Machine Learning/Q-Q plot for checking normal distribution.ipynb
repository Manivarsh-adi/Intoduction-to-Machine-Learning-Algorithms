```python
#Q-Q plot

import numpy as np 
import pylab
import scipy.stats as stats
```

```python
# N(0,1) we are taking a theoritical Quantiles

std_normal=np.random.normal(loc=0,scale=1,size=1000)
print(std_normal.mean())
std_normal.std()
```

```python
# we are finding 100 percentiles of the N(0,1)
for i in range(0,101):
    print(i,np.percentile(std_normal,i))
```

```python
#genrate 100 samples of normal distribution N(20,5) with mean=20,variance=5

measurments=np.random.normal(loc=20,scale=5,size=100)

stats.probplot(measurments,dist='norm',plot=pylab)
pylab.show()
```

**By ploting with 100 samples we can see that points are not so lying on the line**

```python
#tring with 1000 samples
measurments=np.random.normal(loc=20,scale=5,size=1000)
stats.probplot(measurments,dist='norm',plot=pylab)
pylab.show()
```

**when we take 1000 points we can see that almost all points lie on the line**

```python
#tring with less no 50 samples


measurments=np.random.normal(loc=20,scale=5,size=50)
stats.probplot(measurments,dist='norm',plot=pylab)
pylab.show()
```

**When we take 50 points its seems bad than 100 points where q-q plot fails to show the normal distribution**
